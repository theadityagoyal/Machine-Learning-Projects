**Now we will try object oriented style of matplotlib programming**

```python
fig=plt.figure()
```

wee see an instance has been created

**the entire code has to be written together to tie it for execution**

```python
fig=plt.figure()
#add set of axes to it left,bottom,width,height
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')
axes.set_xlabel('x title')
axes.set_ylabel('y title')
axes.set_title('plot title')
```

**Let us create multiple plots in the same canvas**

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main outside axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inside axes

# Larger Figure Axes 1
axes1.plot(x, y, 'r')
axes1.set_xlabel('X_axes1')
axes1.set_ylabel('Y_axes1')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'b')
axes2.set_xlabel('X_axes2')
axes2.set_ylabel('Y_axes2')
axes2.set_title('Axes 2 Title');
```

**we have a powerful bettern option of calling plt.subplots, which is similiar to plt.figure()**

```python
fig, axes = plt.subplots()

# Now we will use the axes object to add elements to the plot
axes.plot(x, y, 'b--')
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title');
```

**we can also create subplots from the same function call**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig    
plt.tight_layout()
```

*Here in the above subplot code, we have used plt.tight_layout to ensure there is no overlapping*

**We can also give size for a plot or subplot**

```python
fig, axes = plt.subplots(figsize=(10,4))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

**Seaborn comes with an inbuilt dataset tips and we will use it here**

```python
#convention to import the seaborn library as sns
import seaborn as sns
#to see visualizations inside notebook
%matplotlib inline
```

```python
tip=sns.load_dataset('tips')
```

```python
type(tip)
```

```python
#to get general idea of data, type of columns
tip.head()
```

**Distplot----
It shows the distribution of single variable's set of observations**

```python
sns.distplot(tip['total_bill'])
```

**Jointplot--
With this, we can match up two distplots for bivariate data. We can specify the kind parameter to compare with**

**i)scatter 
ii)reg
iii)resid
iv)kde
v)hex**

```python
sns.jointplot(x='total_bill',y='tip',data=tip,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip,kind='resid')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip,kind='hex')
```

**rugplot**

# Categorical Plots

**Barplot**

**It aggregates data of a categorical feature in our data.** 

**It is a general plot that allows us to aggregate the categorical data 
based on some function,by default the mean, which can be changed as well**

```python
sns.barplot(x='sex',y='total_bill',data=tip)
```

**countplot**

**It is almost same as barplot**

**Here, the estimator is explicitly counting the number of occurrences** 

**Hence we only pass the single x value**

```python
sns.countplot(x='sex',data=tip)
```

**Boxplot(Box and Whisker plot)**
--

**Used to show the distribution of categorical data.**

**Quantitative data used to show comparisons between variables or across levels of a categorical variable.** 

**The box shows the quartiles of the dataset** 

**The whiskers extend to show the rest of the distribution**

**outliers calculated using a method that is a function of the inter-quartile range.**

```python
sns.boxplot(x="sex", y="total_bill", data=tip,palette='rainbow')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tip, palette="cool")
```

**violinplot**
--

**It is similar to boxplot, but here the components are not mapped to each data point,**

**instead, they represent kernel density estimation of the the underlying distribution**

```python
sns.violinplot(x="day", y="total_bill", data=tip,hue='sex',split=True,palette='coolwarm')
```

**Matrix Plots**
--
**It allows us to plot data as color-encoded matrices** 
**It can also be used to indicate clusters within the data** 

**Types-1)heatmap** and 2)clustermap:

```python
#In order for a heatmap to work properly, the data should be in a matrix form.

#sns.heatmap function just colors it.
```

```python
#if we remember from data analysis that we can check the correlation between columns as->

a=tip.corr()
```

```python
sns.heatmap(a)
```

```python
#to correct the loss of the previous output and add values on the box

ax = sns.heatmap(a, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

**Grids**
--

**PairGrid- It is a subplot grid for plotting pairwise relationships in a dataset.**

```python
g = sns.PairGrid(tip)
g.map(plt.scatter)
```

**pairplot- A simpler version of PairGrid**

```python
sns.pairplot(tip,palette='coolwarm',hue='sex')
```

**Facet Grid-> common way to create grids of plots based on a feature:**

```python
fg = sns.FacetGrid(tip, col="time",  row="smoker")
fg = fg.map(plt.hist, "total_bill")
```

**Regression Plot**
--

```python
sns.lmplot(x='total_bill',y='tip',data=tip,col='day',hue='sex',palette='coolwarm',
          aspect=0.5,size=8)
```
